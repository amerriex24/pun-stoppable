```python
import pandas as pd
```

```python
#create a dataframe for nfl schedule
nfl_sched = "NFLSchedule.csv.zip"
nfl_sched = pd.read_csv(nfl_sched)
nfl_sched
```

```python
#filter for superbowl
super_stat = nfl_sched[nfl_sched['Week'].str.contains("SuperBowl")].copy()
super_stat = super_stat.drop(['YdsL', 'TOL', 'TOW', 'Time', 'YdsW'], axis=1)
super_stat
```

```python
super_stat.sort_values(by=['Year'], inplace=True, ascending=False)
super_stat
```
